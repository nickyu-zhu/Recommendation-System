```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
path_recipe = 'RAW_recipes.csv'
path_interaction = 'RAW_interactions.csv'

data_recipe = pd.read_csv(path_recipe)
data_interaction = pd.read_csv(path_interaction)
```

```python
data_recipe = data_recipe.drop(['contributor_id', 'steps'], axis=1)
```

```python
data_recipe.info()
```

```python
data_recipe.rename(columns={'id': 'recipe_id'}, inplace=True)
```

```python
data_recipe.info()
```

```python
data_interaction.info()
```

```python
data = pd.merge(data_interaction, data_recipe, on="recipe_id", how="left")
```

```python
data.info()
```

```python
data.rename(columns={'date':'review_date', 'submitted': 'recipe_date'}, inplace=True)
```

```python
data.info()
```

```python
data[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = data.nutrition.str.split(",",expand=True) 
```

```python
data['calories'] =  data['calories'].apply(lambda x: x.replace('[','')) 
```

```python
data['carbohydrates (PDV)'] =  data['carbohydrates (PDV)'].apply(lambda x: x.replace(']','')) 
```

```python
data[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = data[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']].astype('float')
```

```python
data.info()
```

```python
data.drop(['nutrition'], axis=1,inplace = True)
```

```python
data['healthy_index'] = data['total fat (PDV)']+data['saturated fat (PDV)']
```

```python
data.head()
```

```python
data.info()
```

```python
data['review_date'] = pd.to_datetime(data['review_date'], format='%Y-%m-%d')
```

```python
data['recipe_date'] = pd.to_datetime(data['recipe_date'], format='%Y-%m-%d')
```

```python
data.loc[(data['saturated fat (PDV)']>50) | (data['sugar (PDV)']>50), 'healthy_index'] = 0 
data.loc[(data['saturated fat (PDV)']<=50) & (data['sugar (PDV)']<=50), 'healthy_index'] = 1
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['name'], axis=1,inplace = True)
```

```python
data.info()
```

```python
data.dropna(subset=['review'], inplace=True)
```

```python
data.info()
```

```python
new_cols = [col for col in data.columns if col != 'rating'] + ['rating']
data = data[new_cols]
```

```python
data.head()
```

```python
data.info()
```

```python
X = data.drop(columns = ['rating']).copy()
y = data['rating']

X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.8)

test_size = 0.5

X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5)
```
